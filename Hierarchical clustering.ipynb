### Creation of Variable MDA8O3:

```python
# creation of necessary datasets to calculate MDA8O3:
FR_26010 = pd.concat([datifr00_26010,datifr01_26010,datifr02_26010,datifr03_26010,datifr04_26010,datifr05_26010,
                      datifr06_26010,datifr07_26010,datifr08_26010,datifr09_26010,datifr10_26010,
                      datifr11_26010,datifr12_26010,], ignore_index=True)

FR_07004 = pd.concat([datifr00_07004, datifr01_07004, datifr02_07004, datifr03_07004, datifr04_07004, datifr05_07004,
                      datifr06_07004, datifr07_07004, datifr08_07004, datifr09_07004, datifr10_07004,
                      datifr12_07004], ignore_index=True)

IE =  pd.concat([ datiIE01, datiIE02, datiIE03, datiIE04, datiIE05,
                      datiIE06,datiIE07, datiIE08, datiIE09, datiIE10, datiIE11,
                      datiIE12], ignore_index=True)

NL = pd.concat([datinl00, datinl01, datinl02, datinl03, datinl04, datinl05,
                      datinl06,datinl07, datinl08, datinl09, datinl10, datinl11,
                      datinl12], ignore_index=True)
GB_first =pd.concat([  datigb00_first, datigb01_first, datigb02_first, datigb03_first, datigb04_first,
               datigb05_first, datigb06_first, datigb07_first, datigb08_first, datigb09_first,
               datigb10_first, datigb11_first, datigb12_first], ignore_index=True)

GB_second =pd.concat([  datigb00_second, datigb01_second, datigb02_second, datigb03_second, datigb04_second,
               datigb05_second, datigb06_second, datigb07_second, datigb08_second, datigb09_second,
               datigb10_second, datigb11_second, datigb12_second], ignore_index=True)

# cycle for calculating MDA8O3 for all the datasets of every station
def process_dataset(dataset):
    # Converting columns with dates into datetime format
    dataset['DatetimeBegin'] = pd.to_datetime(dataset['DatetimeBegin'])
    dataset['DatetimeEnd'] = pd.to_datetime(dataset['DatetimeEnd'])

    # Consider only misurations targeted as valid
    valid_data = dataset[dataset['Validity'] == 1]

    # calculating MOVING AVERAGES 
    valid_data['MDA8O3'] = valid_data['Concentration'].rolling(window=8, min_periods=1).mean()

    # calculating daily max of MDA8O3
    daily_max_MDA8O3 = valid_data.groupby(valid_data['DatetimeBegin'].dt.date)['MDA8O3'].max()

    # Counting number of valid MOVING AVERAGES for every day
    valid_counts = valid_data.groupby(valid_data['DatetimeBegin'].dt.date)['MDA8O3'].count()


    # Filter original dataframe keeping only data of the valid days
    valid_counts.rename('NumH', inplace=True)
    valid_counts = pd.DataFrame(valid_counts)
    daily_max_MDA8O3 = pd.DataFrame(daily_max_MDA8O3)

    merged_data = pd.merge(daily_max_MDA8O3, valid_counts, on='DatetimeBegin', how='inner')

    return merged_data[merged_data['NumH'] >= 18]

datasets = [FR_26010, FR_07004, IE, NL, GB_first, GB_second ]

# Apply the function to each dataset
results = [process_dataset(dataset) for dataset in datasets]

# select each dataset in the list results 
MDA8O3_FR_26010 =pd.DataFrame(results [0])  
MDA8O3_FR_07004 =pd.DataFrame(results [1]) 
MDA8O3_IE =pd.DataFrame(results [2])  
MDA8O3_NL =pd.DataFrame(results [3]) 
MDA8O3_GB_first =pd.DataFrame(results [4])  
MDA8O3_GB_second =pd.DataFrame(results [5]) 


meanMDA_FR_26010= MDA8O3_FR_26010['MDA8O3'].mean()
meanMDA_FR_07004= MDA8O3_FR_07004['MDA8O3'].mean()
meanMDA_IE= MDA8O3_IE['MDA8O3'].mean()
meanMDA_NL= MDA8O3_NL['MDA8O3'].mean()
meanMDA_GB_first= MDA8O3_GB_first['MDA8O3'].mean()
meanMDA_GB_second= MDA8O3_GB_second['MDA8O3'].mean()
```
